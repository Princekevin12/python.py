```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def is_circular_prime(n):
    s = str(n)
    for i in range(len(s)):
        r = int(s[i:] + s[:i])
        if not is_prime(r):
            return False
    return True

n = int(input("Enter number: "))

if is_circular_prime(n):
    print("Circular Prime")
else:
    print("Not Circular Prime")
```

### Finding Prime Numbers whose Reverse is also Prime

To achieve this, we need two helper functions:
1.  `is_prime(num)`: Checks if a given number is prime.
2.  `reverse_number(num)`: Reverses the digits of a given number.

```python
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print("Testing is_prime:")
print(f"is_prime(7): {is_prime(7)}")
print(f"is_prime(10): {is_prime(10)}")
```

```python
def reverse_number(num):
    return int(str(num)[::-1])

print("\nTesting reverse_number:")
print(f"reverse_number(123): {reverse_number(123)}")
print(f"reverse_number(701): {reverse_number(701)}")
```

Now, let's find the numbers that satisfy the condition (both the number and its reverse are prime).

```python
prime_and_reversed_prime = []
max_num_to_check = 1000 # You can adjust this limit

print(f"\nSearching for numbers up to {max_num_to_check} where both the number and its reverse are prime:")

for num in range(2, max_num_to_check + 1):
    if is_prime(num):
        reversed_num = reverse_number(num)
        # Avoid duplicates for palindromic primes (e.g., 11, 101)
        # Also ensure the reversed number is not the same as original unless it's a palindromic prime
        if is_prime(reversed_num) and num <= reversed_num:
            prime_and_reversed_prime.append((num, reversed_num))

for original, reversed_val in prime_and_reversed_prime:
    print(f"Original: {original}, Reversed: {reversed_val}")

if not prime_and_reversed_prime:
    print("No such numbers found within the specified range.")
```
